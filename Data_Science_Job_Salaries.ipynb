## Analyze Salary Distribution

### Subtask:
Generate a histogram or box plot to visualize the overall distribution of salaries, including a legend.


**Reasoning**:
To visualize the distribution of salaries as a histogram, I need to import the `matplotlib.pyplot` library and then use its `hist` function on the `salary_in_usd` column of the DataFrame. I will also add appropriate labels and a title for clarity, as specified in the instructions.



```python
import matplotlib.pyplot as plt

# Create a histogram of the salary_in_usd column
plt.figure(figsize=(10, 6))
plt.hist(df['salary_in_usd'], bins=50,color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of Salaries (USD)')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Analyze Salary by Experience Level

### Subtask:
Investigate how salaries vary across different experience levels (e.g., Entry-level, Mid-level, Senior-level, Executive-level). Visualize this relationship using an appropriate plot (e.g., bar plot or box plot) with a legend.


### Reasoning:
To visualize how salaries vary across different experience levels, a box plot is a suitable choice as it effectively displays the distribution (median, quartiles, and outliers) of `salary_in_usd` for each `experience_level`. I will use the `seaborn` library for its enhanced plotting capabilities, along with `matplotlib.pyplot` for plot customization like titles and labels.

**Reasoning**:
To visualize the salary distribution across different experience levels, I will generate a box plot using the `seaborn` library. This plot will clearly show the median, quartiles, and potential outliers for `salary_in_usd` grouped by `experience_level`, and will be customized with appropriate labels and a title for readability.



```python
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Experience Level (USD)')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

## Analyze Salary by Job Title

### Subtask:
Explore salary variations for different job titles. Identify and visualize the top 10 highest-paying job titles using a bar plot, including a legend.


**Reasoning**:
To identify the top 10 highest-paying job titles, I will group the DataFrame by 'job_title', calculate the mean of 'salary_in_usd' for each, sort them in descending order, and then select the top 10. Following that, I will create a bar plot to visualize these results.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average salary for each job title
average_salaries_by_job_title = df.groupby('job_title')['salary_in_usd'].mean()

# Sort in descending order and get the top 10
top_10_job_titles = average_salaries_by_job_title.sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x=top_10_job_titles.index, y=top_10_job_titles.values, palette='viridis')
plt.title('Top 10 Highest Paying Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.grid(axis='y', alpha=0.75)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

**Reasoning**:
To resolve the `FutureWarning` from seaborn regarding the `palette` argument, I will explicitly assign the `x` variable to the `hue` parameter and set `legend=False` in the `sns.barplot` function, as suggested by the warning.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average salary for each job title
average_salaries_by_job_title = df.groupby('job_title')['salary_in_usd'].mean()

# Sort in descending order and get the top 10
top_10_job_titles = average_salaries_by_job_title.sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x=top_10_job_titles.index, y=top_10_job_titles.values, hue=top_10_job_titles.index, palette='viridis', legend=False)
plt.title('Top 10 Highest Paying Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.grid(axis='y', alpha=0.75)
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
