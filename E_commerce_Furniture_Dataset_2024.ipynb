## Load Data

### Subtask:
Load the 'WA_Fn-UseC_-HR-Employee-Attrition.csv' file into a pandas DataFrame to begin our analysis.


**Reasoning**:
To load the data as instructed, I will first import the pandas library and then use its `read_csv` function to load the specified CSV file into a DataFrame named `df`.



```python
import pandas as pd
df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')
print("DataFrame 'df' loaded successfully. Displaying the first 5 rows:")
df.head()
```

## Initial Data Exploration

### Subtask:
Perform an initial exploration of the dataset. This includes displaying the first few rows, checking data types, and reviewing descriptive statistics to get an overview of the data's structure and content.


**Reasoning**:
To perform the initial data exploration, I will display the first few rows, check the data types and non-null values, and generate descriptive statistics for numerical columns using `head()`, `info()`, and `describe()` methods respectively.



```python
print("Displaying the first 5 rows of the DataFrame:")
df.head()

print("\nConcise summary of the DataFrame (data types and non-null values):")
df.info()

print("\nDescriptive statistics for numerical columns:")
df.describe()
```

## Analyze Attrition Distribution

### Subtask:
Visualize the overall distribution of the 'Attrition' column to understand the proportion of employees who have attrited. This will be represented as a bar plot with appropriate labels and a clear title, including a legend if applicable.


**Reasoning**:
To visualize the overall distribution of the 'Attrition' column, I need to import the necessary plotting libraries and then create a count plot using seaborn, adding appropriate labels and a title.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot of the 'Attrition' column
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Attrition', hue='Attrition', palette='viridis', legend=False)

# Add title and labels
plt.title('Distribution of Attrition')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')

# Display the plot
plt.show()
```

## Explore Key Factors Related to Attrition

### Subtask:
Investigate the relationship between 'Attrition' and key features like 'Department' and 'Age' using appropriate visualizations.


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Department', hue='Attrition', palette='viridis')
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Attrition', y='Age', palette='viridis')
plt.title('Attrition by Age Group')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()
```

## Explore Key Factors Related to Attrition

### Subtask:
Investigate the relationship between 'Attrition' and the 'JobRole' feature using a bar chart.


```python
plt.figure(figsize=(14, 7))
sns.countplot(data=df, x='JobRole', hue='Attrition', palette='viridis')
plt.title('Attrition by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.legend(title='Attrition')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

## Explore Key Factors Related to Attrition

### Subtask:
Investigate the relationship between 'Attrition' and the 'EducationField' feature using a bar chart.


**Reasoning**:
To investigate the relationship between 'Attrition' and 'EducationField', I will create a count plot (bar chart) as specified. This will show the number of employees who attrited or not, categorized by their education field.



```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='EducationField', hue='Attrition', palette='viridis')
plt.title('Attrition by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.legend(title='Attrition')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

Investigate the relationship between 'Attrition' and the 'MonthlyIncome' feature using a box plot.


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', hue='Attrition', palette='viridis', legend=False)
plt.title('Attrition by Monthly Income')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```
